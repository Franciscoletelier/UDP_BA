```python
# Llamamos las librerías correspondientes
import pandas as pd
import numpy as np
```

## Creación del df

```python
# Creamos nuestro objeto 'df' para el ejemplo:
df = pd.DataFrame()

# Es un objeto vacío hasta el momento
```

```python
# Creamos las columnas y datos que contienen las columnas
# Son datos creados random entre un min= 0 y un max que yo determino.
# No tiene que ser igual para todos.

df['Ventas'] = pd.DataFrame(np.random.randint(low=0, high=15000000, size=50, dtype=int))
df['Costos'] = pd.DataFrame(np.random.randint(low=0, high=10000000, size=50, dtype=int))
df['Honorarios'] = pd.DataFrame(np.random.randint(low=0, high=9000000, size=50, dtype=int))
df['RRHH'] = pd.DataFrame(np.random.randint(low=0, high=5000000, size=50, dtype=int))
df['Gastos'] = pd.DataFrame(np.random.randint(low=0, high=1500000, size=50, dtype=int))
df['ID_cliente'] = pd.DataFrame(np.random.randint(low=0, high=10, size=50, dtype=int))

print(df)
```

```python
# Ejemplo

df_ = pd.DataFrame(np.random.rand(10,2),columns = ['columna1', 'columna2'],dtype=float) # Podemos especificar el tipo de variable, pero debe ser todo de df

print(df_[['columna1','columna2']])

# Si se fija, el uso del doble corchete es otra forma de llamara las variables
```

```python
# Vamos a agregar una columna con datos faltantes y una columna vacía
import random
df['productos'] = pd.DataFrame([random.randint(0,5) if random.randint(0, 1) else np.nan for i in range(len(df))]).astype('object')
```

## Funcion iloc y loc para seleccionar multiple columnas
 * Usando la sintaxis __getitem__ ([]) para seleccionar múltiples columnas
 * Usando los métodos iloc() y loc() para seleccionar múltiples columnas

```python
# Quiero revisar la columna de Ventas y Costos con getitem
df[['Ventas', 'Costos']].head(10)
```

```python
# Quiero revisar la columna Venta y Costos con Loc
#[filas desde: fila hasta,['columna1', 'columna2']]
df.loc[10:15,['Ventas', 'Costos']]
# Con el primer corchete puedo limitar las filas a mostrar o ubicar
```

```python
# Cuando queramos usar los índices de las columnas para extraerlas, podemos usar iloc()
#[filas desde: fila hasta,['numero de la columna', 'numero de la columna']]
df.iloc[15:25,[2,6]]
# Con el primer corchete puedo limitar las filas a mostrar o ubicar
```

### Función copy() y drop()

```python
# Se usa para copiar df o parte de el.
# Es una buena práctica

df1 = df.copy()
df1.head(10)
```

```python
# La función drop se usa para eliminar columnas

df1 = df1.drop(['ID_cliente'], axis=1)
df1.head()

# https://pandas.pydata.org/pandas-docs/stable/reference/api/pandas.DataFrame.drop.html?highlight=drop#pandas.DataFrame.drop
```

```python
# La función drop se usa para eliminar filas

df1 = df1.drop([0,2,4])
df1.head()
```

### Funcion isnull e isnan

```python
# Para verificar el numero de datos nulos
df.isnull().sum(axis=0)
```

```python
# Condición booleana para verificar el numero de datos Na pero con una sola variable/columna

df['productos'].isna().sum()
```
