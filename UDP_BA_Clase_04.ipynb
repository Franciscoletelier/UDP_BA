# Condicional IF ELSE

* Aspectos básicos
  * El condicional ***if-else*** es fundamental en la programación, ya que te permite ejecutar diferentes bloques de código dependiendo de si una condición es verdadera o falsa (booleana)

```python
if condicion:
    # Aquí va el bloque de código si la condición es verdadera
else:
    # Y acá el código si la condición es falsa
```

### Ejemplo 1: ingresar un número:
* En este ejemplo, el programa solicita al usuario un número, y luego utiliza la declaración if-else para determinar si es positivo, negativo o cero.

* Recuerda que:
  * Puedes tener múltiples bloques elif para manejar más de dos casos.
  * El bloque else es opcional y se ejecutará si ninguna de las condiciones anteriores es verdadera.


```python
numero = float(input("Ingrese un número: "))

if numero > 0:
    print("El número es positivo")
elif numero < 0:
    print("El número es negativo")
else:
    print("El número es cero")


# En este caso, la condición es que el número
# sea mayo a cero, o menor a cero
```

# Ciclo for + If Else

### Ejemplo 01: for + If else

* En este ejemplo, utilizamos un ciclo for para iterar a través de un rango de números y utilizamos la declaración if-else para imprimir si cada número es par o impar:

```python
for numero in range(1, 11):
    if numero % 2 == 0:
        print(f"{numero} es un número par.")
    else:
        print(f"{numero} es un número impar.")
```

### Ejemplo 2: Suma de los primeros números naturales pares en un rango
* En este ejemplo, usamos un ciclo for para sumar los primeros números naturales pares dentro de un rango y utilizamos la declaración if para verificar si un número es par:

```python
suma = 0
for numero in range(1, 11):
    if numero % 2 == 0:
        suma += numero
print(f"La suma de los primeros números pares es: {suma}")
```

### Ejemplo 3: Conteo de vocales en una cadena
* En este ejemplo, utilizamos un ciclo for para recorrer una cadena y la declaración if-else para contar la cantidad de vocales presentes:

```python
cadena = "Hola, cómo estás?"
contador_vocales = 0

for caracter in cadena:
    if caracter.lower() in "aeiou":
        contador_vocales += 1

print(f"La cadena tiene {contador_vocales} vocales.")
```

### Ejemplo 4: Generador de tablas de multiplicar personalizadas
* En este ejemplo, usamos un ciclo for y la declaración if-else para generar una tabla de multiplicar personalizada para un número ingresado por el usuario:

```python
numero = int(input("Ingresa un número para generar su tabla de multiplicar: "))

print(f"Tabla de multiplicar del {numero}:")
for i in range(1, 11):
    resultado = numero * i
    print(f"{numero} x {i} = {resultado}")
```

### Ejemplo 5. Aplicación para determinar promedio y asignar un clasificador

```python
# Crear una lista de calificaciones
calificaciones = [85, 92, 78, 90, 88, 95, 80, 85, 70, 93]

# Calcular el promedio
total_calificaciones = sum(calificaciones)
cantidad_calificaciones = len(calificaciones)
#promedio = total_calificaciones / cantidad_calificaciones
prom_notas = np.average(calificaciones)

# Imprimir el promedio
print(f"El promedio de calificaciones es: {prom_notas:.2f}")

# Asignar una letra según la clasificación
for calificacion in calificaciones:
    if calificacion >= 90:
        letra = "A"
    elif calificacion >= 80:
        letra = "B"
    elif calificacion >= 70:
        letra = "C"
    elif calificacion >= 60:
        letra = "D"
    else:
        letra = "F"

    print(f"Calificación: {calificacion} - Clasificación: {letra}")
```

### Ejemplo Bonus: crear una columna, calcular el valor a partir de otra columna con valores

```python
import pandas as pd

# Crear el DataFrame de ejemplo
data = {'Condición': ['A', 'B', 'A', 'B', 'A'],
        'Valor': [10, 20, 30, 40, 50]}

df = pd.DataFrame(data)
df
```

```python
# Crear una nueva columna para el porcentaje
df['Porcentaje'] = 0  # Inicializar la columna con ceros

# Calcular el porcentaje usando un ciclo for y if-else
for index, row in df.iterrows():
    if row['Condición'] == 'A':
        df.at[index, 'Porcentaje'] = row['Valor'] * 0.3
    elif row['Condición'] == 'B':
        df.at[index, 'Porcentaje'] = row['Valor'] * 0.7

# Imprimir el DataFrame resultante
print(df)
```
